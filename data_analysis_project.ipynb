When reviewing the features, we notice some inconsistencies—for example, certain **categorical values appear in numerical columns**, and in some cases, **numerical values are mistakenly found in categorical columns**. This type of mismatch is a common data quality issue.

Another inconsistency involves the presence of **floating-point numbers in columns that should contain only discrete (whole number) values**, such as counts of internships or certifications.

To address these inconsistencies, we will apply appropriate data cleaning techniques to handle unexpected or incorrect values in each column. Specifically:
- If a **categorical value appears in a numerical column**, we will replace it with the **mean (average)** of the valid numerical entries.
- If a **numerical value appears in a categorical column**, we will replace it with the **mode** (most frequently occurring value) of that column.

These steps will help standardize the dataset and prepare it for accurate analysis and modeling.

```python
Inconsistent_Ages = df[df['Age'].apply(lambda x: not isinstance(x, int) or (isinstance(x, (int, float)) and x < 18) or (isinstance(x, (int, float)) and x > 30))]


Unique_Inconsistent_Ages = Inconsistent_Ages['Age'].unique()
print("\nUnique Inconsistent values in the 'Age' column:")
print(Unique_Inconsistent_Ages)
print("_"*100)

df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

Rounded_meanAge_value = int(round(df['Age'].mean()))

print(f"The average age is {Rounded_meanAge_value}")

df.loc[(df["Age"] < 0) | (df["Age"] ==8) , "Age"] = Rounded_meanAge_value # the negative value -1 and the value 8  do not seem logical or a typo so it was replaced by the rounded average age
df.loc[(df['Age'] == 200) , 'Age'] = 20 # the value 200 seems to be a typo so instead of 200 it has been written 20.
df["Age"].fillna(Rounded_meanAge_value, inplace=True) # Nan values were replaced by the rounded average age.

df["Age"] = df["Age"].apply(
    lambda x: int(x * 100) if isinstance(x, float) and (x < 1)
    else int(x) if isinstance(x, (int, float))
    else int((x*10)) if not isinstance(x, int) and x >= 1 
    else (x*10) if isinstance(x,int)<10
    else x
)

display(df['Age'])

df["Age"].info()
```

The string value "Law" has been replaced by Nan, and then the Nan values will be replaced by the rounded average

The float values has been fixed by multiplying values that are less than 1 by 100 and the values that are more than 1 by 10. Also, the type of the column has been fixed to integer while keeping the missing values and outliers same as before.

For the case of the value 200, it seems to be a typo so it was fixed by replacing it by its logical replacement 20. 
For the case of the values -1 and 8, they do not seem to be typo so they were fixed by replacing them by the rounded average to keep the consistency of the column type.

**Why this is important:**

### Making Data Consistent for Our Model

This code makes sure all our training and testing data is on the same scale.

**What it does:**

1.  **Brings Everything to the Same Level:** It uses a tool called `StandardScaler` to adjust the numbers in our data so they all have an average of zero and a standard spread. This includes both the factors we use to predict salary and the salary itself.

2.  **Learns from the Training Data:** It first looks at our training data to figure out the average and spread of each factor and the salary.

3.  **Applies the Learning:** Then, it uses what it learned from the training data to adjust the numbers in both the training data and the separate testing data. This ensures both sets are treated the same way.

**Why this is important:**

Some computer learning methods work better when all the numbers are on a similar scale. This process prevents factors with larger numbers from unfairly influencing the learning process.

```python
from sklearn.metrics import mean_squared_error

## OLS
from sklearn.linear_model import LinearRegression
reg1 = LinearRegression(fit_intercept=False).fit(X_train, y_train)
y_pred1 = reg1.predict(X_test)
print('The MSE using OLS is:', mean_squared_error(y_test, y_pred1))


## Ridge
from sklearn.linear_model import RidgeCV
reg2 = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3], fit_intercept=False,cv=10).fit(X_train, y_train)
y_pred2 = reg2.predict(X_test)
print('The MSE using Ridge is:', mean_squared_error(y_test, y_pred2))


## Lasso
from sklearn.linear_model import LassoCV
reg3 = LassoCV(alphas=[1e-3, 1e-2, 1e-1, 1e0, 1e1, 1e2, 1e3], 
               fit_intercept=False,cv=10, random_state=0).fit(X_train, y_train)
y_pred3 = reg3.predict(X_test)
print('The MSE using Lasso is:', mean_squared_error(y_test, y_pred3))
```

### Comparing Models for Salary Prediction

This code trains and checks three ways to predict starting salary: Ordinary Least Squares (OLS), Ridge Regression, and Lasso Regression.

**How it works:**

* **Checks Model Accuracy:** It uses "Mean Squared Error" (MSE) to see how wrong the predictions are. Lower MSE is better.
* **OLS:** A basic way to draw a line that best fits the training data.
* **Ridge:** Like OLS, but it also tries to keep the line simple to avoid being too specific to the training data. It automatically finds the best level of simplicity.
* **Lasso:** Similar to Ridge, but it can completely ignore some factors if they don't seem important. It also finds the best level of simplicity automatically.

**What the results mean:**

Based on these results, **OLS (Ordinary Least Squares) had the lowest Mean Squared Error** on the test data. This suggests that, for this particular dataset and these models, the basic OLS model provided slightly more accurate predictions on unseen data compared to Ridge and Lasso. The regularization techniques employed by Ridge and Lasso, which aim to prevent overfitting by adding a penalty to the model's complexity, did not lead to improved performance on the test set in this instance. **Why is it so low?**

A low R-squared can happen for several reasons:

* **The relationship between the factors and starting salary might be largely non-linear.** Our regression model assumes a straight-line relationship, but the actual relationship could be much more complex.
* **There is likely a significant amount of random variation or "noise" in the starting salary data** that our model cannot account for.
